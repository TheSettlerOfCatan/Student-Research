```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
test1 = np.loadtxt('Downloads/test1.csv', delimiter=', ')
test2 = np.loadtxt('Downloads/test2.csv', delimiter=', ')
```

```python
plt.subplots(figsize=(15,8))
plt.subplot(2,2,1)
plt.plot(test1[:,0],test1[:,1])

plt.subplot(2,2,2)
plt.plot(test2[:,0],test2[:,1])

plt.subplot(2,2,3)
plt.plot(test1[:,0],test1[:,2])

plt.subplot(2,2,4)
plt.plot(test2[:,0],test2[:,2])
plt.show()
```

```python
plt.subplots(figsize=(12,8))
plt.subplot(2,1,1)
plt.plot(test1[:,0],test1[:,1])
plt.plot(test2[:,0],test2[:,1])

plt.subplot(2,1,2)
plt.plot(test1[:,0],test1[:,2])
plt.plot(test2[:,0],test2[:,2])
plt.show()
```

```python
plt.subplots(figsize=(15,8))
plt.subplot(2,2,1)
plt.plot(test1[(np.argmax(test1[:,0]) + 1):np.argmin(test1[:,0]), 0], test1[(np.argmax(test1[:,0]) + 1):np.argmin(test1[:,0]), 1])

plt.subplot(2,2,2)
plt.plot(test2[(np.argmax(test2[:,0]) + 1):np.argmin(test2[:,0]), 0], test2[(np.argmax(test2[:,0]) + 1):np.argmin(test2[:,0]), 1])

plt.subplot(2,2,3)
plt.plot(test1[(np.argmax(test1[:,0]) + 1):np.argmin(test1[:,0]), 0], test1[(np.argmax(test1[:,0]) + 1):np.argmin(test1[:,0]), 2])

plt.subplot(2,2,4)
plt.plot(test2[(np.argmax(test2[:,0]) + 1):np.argmin(test2[:,0]), 0], test2[(np.argmax(test2[:,0]) + 1):np.argmin(test2[:,0]), 2])
plt.show()
```

```python
plt.subplots(figsize=(12,8))
plt.subplot(2,1,1)
plt.plot(test1[(np.argmax(test1[:,0]) + 1):np.argmin(test1[:,0]), 0], test1[(np.argmax(test1[:,0]) + 1):np.argmin(test1[:,0]), 1])
plt.plot(test2[(np.argmax(test2[:,0]) + 1):np.argmin(test2[:,0]), 0], test2[(np.argmax(test2[:,0]) + 1):np.argmin(test2[:,0]), 1])

plt.subplot(2,1,2)
plt.plot(test1[(np.argmax(test1[:,0]) + 1):np.argmin(test1[:,0]), 0], test1[(np.argmax(test1[:,0]) + 1):np.argmin(test1[:,0]), 2])
plt.plot(test2[(np.argmax(test2[:,0]) + 1):np.argmin(test2[:,0]), 0], test2[(np.argmax(test2[:,0]) + 1):np.argmin(test2[:,0]), 2])
plt.show()
```
