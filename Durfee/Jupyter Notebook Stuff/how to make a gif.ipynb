```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
x = np.arange(0,N,0.1)
images = []
for t in range(N):
    
    figure1 = plt.figure(figsize=(10,5))
#     plt.subplots(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.plot(x,np.sin(x/4+t/2))
    plt.axis([0,101,-1.1,1.1])
    
    plt.subplot(1,2,2)
    plt.plot(x,np.cos(x/4+t/2))
    plt.axis([0,101,-1.1,1.1])
    plt.tight_layout()
    
    figure1.savefig("/home/carter/Documents/Research/GIF testing/plotfig.png")
    plt.close(figure1)
    
    images.append(imageio.imread("/home/carter/Documents/Research/GIF testing/plotfig.png"))
#     image.append(plt.plot(x,np.sin(x*t)))
imageio.mimsave('/home/carter/Documents/Research/GIF testing/gifresearch2.gif', images, duration=0.15)
# image
# xx, sinx = np.meshgrid(x,np.sin(x))
# plt.imshow(sinx)
# plt.show()
```
